### 游늸 Extrair a correla칞칚o entre raridade e taxa de vit칩ria.

```python
# TRANSFORMAR AS RARIDADES EM DADOS N칔MERICOS, PARA QUE SE POSSA FAZER A CORRELACAO ENTRE ELA E O WIN RATE
# CADA RARIDADE RECEBER츼 UM N칔MERO QUE VAI O IDENTIFICAR

# lista que cont칠m todas as raridades das cartas
lista_raridades = df_cartas_copias['rarity']

dic_raridades_numericas = {
    'common': 1,
    'rare': 2,
    'epic': 3,
    'legendary': 4,
    'champion': 5
}

# funcao que retorna a raridade agora em n칰merico
def transformar_raridade_numerico(valor):
    valor = dic_raridades_numericas[valor]
    return valor

# lista com as raridades mas agora em numericos

lista_raridades_numericas = lista_raridades.map(transformar_raridade_numerico)
```

```python
# criar uma c칩pia do df original, somente para an치lise da correlacao entre raridade e win rate, para que nao fique ru칤dos no df
df_correlacao = df_cartas.copy()
```
