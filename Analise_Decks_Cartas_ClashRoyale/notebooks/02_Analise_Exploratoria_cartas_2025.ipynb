```python
# IMPORTANDO A BASE DE DADOS
df_cartas = pd.read_excel('../data/raw/clash_royale_cards.xlsx')
```

```python
# CRIANDO A C√ìPIA DO DATAFRAME PARA SE FAZER AS AN√ÅLISES
df_cartas_copias = df_cartas.copy()
```

### üìç Extrair as top 10 cartas com a maior taxa de vit√≥ria

```python
# CRIANDO O DF COM O TOP 10 CARTAS COM O MAIOR WIN RATE (taxa de vit√≥ria)
df_top10_cartas_vitoriosas_2025 = df_cartas_copias.sort_values(by='Win Rate', ascending=False).head(10)
df_top10_cartas_vitoriosas_2025
```

### üìç Extrair a correla√ß√£o entre raridade e taxa de vit√≥ria.

```python
# TRANSFORMAR AS RARIDADES EM DADOS N√öMERICOS, PARA QUE SE POSSA FAZER A CORRELA√á√ÉO ENTRE ELA E O WIN RATE
# CADA RARIDADE RECEBER√Å UM N√öMERO QUE VAI O IDENTIFICAR
# lista que cont√©m todas as raridades das cartas
lista_raridades = df_cartas_copias['rarity']

dic_raridades_numericas = {
    'common': 1,
    'rare': 2,
    'epic': 3,
    'legendary': 4,
    'champion': 5
}
# funcao que retorna a raridade agora em n√∫merico
def transformar_raridade_numerico(valor):
    valor = dic_raridades_numericas[valor]
    return valor

# lista com as raridades mas agora em numericos
lista_raridades_numericas = lista_raridades.map(transformar_raridade_numerico)
```

```python
# criar uma c√≥pia do df original, somente para an√°lise da correlacao entre raridade e win rate, para que nao fique ru√≠dos no df
df_correlacao = df_cartas.copy()
# Mudando o valor da coluna rarity para a lista n√∫merica
df_correlacao['rarity'] = lista_raridades_numericas

# Extrair a correla√ß√£o real entre a raridade e win rate
df_valor_correlacao = df_correlacao[['rarity', 'Win Rate']].corr()
df_valor_correlacao
```

```python
# Tirar a m√©dia de win rate entre as raridade para analisar junto com a correla√ß√£o e verificar as tend√™ncias de cartas no meta
grupo_raridade = df_cartas_copias.groupby('rarity')
grupo = grupo_raridade.describe()['Win Rate']['mean'].to_dict()
valor_medias_raridade = transformar_dict_em_df(grupo, 'M√©dia Win Rate por Raridade')
valor_medias_raridade['M√©dia Win Rate por Raridade'] = valor_medias_raridade['M√©dia Win Rate por Raridade'].map('{:.2f}'.format)
valor_medias_raridade
```
